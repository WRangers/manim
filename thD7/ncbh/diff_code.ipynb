```python
from scipy.integrate import odeint
# 为了方便绘图,调用scipy.integrate.odeint而不用sympy.dsolve
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 微分方程组
def lorzen(w, t, sigma, pho, beta):
    # w为初值向量
    # t为时间
    # dx/dt, dy/dt, dz/dt初值
    x, y, z = w
    # 与Lorenz的计算公式对应
    return np.array([sigma*(y-x), x*(pho-z)-y, x*y-beta*z])

t = np.arange(0, 30, 0.01) # 创建时间点
# 调用ode对lorenz进行求解,用两个不同的初始值
track1 = odeint(lorzen, (10.0,8/3,28),t,args=(10.0,28.0,3.0))
track2 = odeint(lorzen,(0.0,1.01,0.0),t,args=(10.0,28.0,3.0))

# 绘图
fig=plt.figure()
ax=Axes3D(fig)
ax.plot(track1[:,0],track1[:,1],track1[:,2])
ax.plot(track2[:,0],track2[:,1],track2[:,2])
plt.show()
```
