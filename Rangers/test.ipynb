```python
from scipy import fftpack
b = 30
f_s = 800
N = 8000
t = np.linspace(-2*np.pi, 2*np.pi, N, endpoint=False)
sq = abs(np.sin(t))

F = fftpack.fft(sq)
f = fftpack.fftfreq(N, 1.0/f_s)

print(F)
print(f)

F_filtered = F * (abs(f) < 5)

ift = fftpack.ifft(F[1:5])
mask = np.where(f >= 0)

fig, axes = plt.subplots(4,1)
axes[0].plot(t, sq)
axes[0].set_ylim(-2, 2)
axes[1].plot(f[mask], abs(F[mask])/N, label="freq")
axes[2].plot(t,ift.real, label="all")
axes[3].plot(t,ift.real, label="zoom")
axes[3].set_xlim(-7, 7)

plt.show()
```
